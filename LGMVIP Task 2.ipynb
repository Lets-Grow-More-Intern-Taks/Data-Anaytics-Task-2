<h2>TASK 2 - Image to Pencil Sketch</h2>

<h3>DATA AUDIT</h3>

<h4>Importing Libraries</h4>

```python
import cv2
import matplotlib.pyplot as plt
```

<h4>Reading Original Image</h4>

```python
image = cv2.imread("Puppy.jpg")
plt.imshow(image[:,:,::-1])
```

<h4>Converting original image to grayscale</h4>

```python
gray_image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image)
```

<h4>Inverting the grayscale image</h4>

```python
inverted_image = cv2.bitwise_not(gray_image)
plt.imshow(inverted_image)
```

<h4>Blurring the image using Gaussian function</h4>

```python
blurred = cv2.GaussianBlur(inverted_image, (111,111),0)
plt.imshow(blurred)
```

<h4>Inverting the blurred image</h4>

```python
inverted_blurred=cv2.bitwise_not(blurred)
plt.imshow(inverted_blurred)
```

<h3>Pencil Sketch</h3>

```python
sketch_image=cv2.divide(gray_image,inverted_blurred, scale=256.0)
pencil_sketch=cv2.cvtColor(sketch_image, cv2.COLOR_BGR2RGB)
plt.imshow(pencil_sketch)
plt.axis('off')
plt.show()
```
